```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df2022 = pd.read_csv('data2022.csv')




df2022 = df2022.drop(index=[3,9,20,26,33,41,48])

df2022 = df2022.reset_index()

df2022
```

```python
plt.plot(df2022['Week'], df2022['Top 10 Gross'])
plt.xlabel("Week Number")
plt.ylabel("Gross (100 million $)")
plt.title("2022 Gross Sales by week")
plt.savefig("gross2022.pdf", bbox_inches="tight")
plt.show()
```
