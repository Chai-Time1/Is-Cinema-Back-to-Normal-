```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df2020 = pd.read_csv('data2020.csv')




df2020 = df2020.drop(index=[5,6,14,20,36,43,52,56])

df2020 = df2020.reset_index()

df2020
```

```python
plt.plot(df2020['Week'], df2020['Top 10 Gross'])
plt.xlabel("Week Number")
plt.ylabel("Gross (100 million $)")
plt.title("2020 Gross Sales by week")
plt.savefig("gross2020.pdf", bbox_inches="tight")
plt.show()
```
