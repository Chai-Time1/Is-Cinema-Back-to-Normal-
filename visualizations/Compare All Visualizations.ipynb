```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df2015_2019 = pd.read_csv('Avgof2015to2019.csv')
df2015_2019
```

```python
df2020 = pd.read_csv('data2020.csv')




df2020 = df2020.drop(index=[5,6,14,20,36,43,52,56])

df2020 = df2020.reset_index()

df2020
```

```python
df2021 = pd.read_csv('data2021.csv')

df2021 = df2021.drop(index=[6,7,15,21,30,37,46,54,59])

df2021 = df2021.reset_index()

df2021
```

```python
df2022 = pd.read_csv('data2022.csv')




df2022 = df2022.drop(index=[3,9,20,26,33,41,48])

df2022 = df2022.reset_index()

df2022
```

```python
plt.plot(df2015_2019['Week'], df2015_2019['mean'], label='2015-2019')
plt.plot(df2020['Week'], df2020['Top 10 Gross'], label='2020')
plt.plot(df2021['Week'], df2021['Top 10 Gross'], label='2021')
plt.plot(df2022['Week'], df2022['Top 10 Gross'], label='2022')
plt.xlabel("Week Number")
plt.ylabel("Gross (100 million $)")
plt.title("Gross Sales by week")
plt.legend()
plt.savefig("grossCombined.pdf", bbox_inches="tight")
plt.show()
```
