```python
url2015 = "https://www.boxofficemojo.com/weekend/by-year/2015/"
url2015
```

```python
import pandas as pd
df = pd.read_html(url2015, header = 0)
df
```

```python
df2015 = df[0]
df[0]
```

# Data Cleaning

```python
df2015.drop('Genre', inplace=True, axis=1)
```

```python
df2015.drop('Budget', inplace=True, axis=1)
```

```python
df2015.drop('Running Time', inplace=True, axis=1)
```

```python
df2015['Top 10 Gross'] = df2015['Top 10 Gross'].str.replace('\W', '')
```

```python
df2015.info()
```

```python
df2015['Top 10 Gross'] = df2015['Top 10 Gross'].astype('int64')
```

# Visualize

```python
import matplotlib.pyplot as plt
```

```python
# Plotted 2015 data Top Grosses throught year

plt.plot(df2015['Week'],df2015['Top 10 Gross'])
plt.show()
```

```python
# Save dataframe to csv

# saving the dataframe
df2015.to_csv('data2015.csv')
```
