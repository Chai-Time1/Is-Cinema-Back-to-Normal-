```python
url2015 = "https://www.boxofficemojo.com/weekend/by-year/2015/"
url2015
```

```python
import pandas as pd
df = pd.read_html(url2015, header = 0)
df
```

```python
df2015 = df[0]
df[0]
```

```python
import seaborn as sns
```

```python
sns.lineplot(data = df2015, x = 'Week', y = 'Releases')
```

```python
sns.distplot(df2015.Releases, 
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2))
```

```python
sns.distplot(df2015.Releases,
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2),
             color='#00BFC4')
```
