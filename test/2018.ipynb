```python
url2018 = "https://www.boxofficemojo.com/weekend/by-year/2018/"
url2018
```

```python
import pandas as pd
df = pd.read_html(url2018, header = 0)
df
```

```python
df2018 = df[0]
df2018
```

```python
df2018.drop('Genre', inplace = True, axis = 1)
df2018.drop('Running Time', inplace = True, axis = 1)
df2018.drop('Budget', inplace = True, axis = 1)
df2018
```

```python
df2018['Top 10 Gross'] = df2018['Top 10 Gross'].str.replace('\W', '')
df2018['Top 10 Gross'] = df2018['Top 10 Gross'].astype('int64')
df2018.info()
df2018
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df2018['Week'], df2018['Top 10 Gross'])
plt.title("2018 Top 10 Gross For Each Week")
plt.xlabel("Week")
plt.ylabel("Top 10 Gross (in $100 millions)")
plt.show()
# plotted a line graph with Week vs. Top 10 Gross amount
```

```python
# Save dataframe to csv

df2018.to_csv('data2018.csv')
```
