## **URL to the webpage "Box Offce Data from Box Office Mojo"**

### Method 1

We could modularize the URL and year strings of text that may be useful for programmatic access (in the subsequent code cell).

```python
# The Building Blocks
url = 'https://www.boxofficemojo.com/weekend/by-year/2019/'

# Combining the URL + year strings together
url
```

## **Read HTML webpage into pandas**

```python
import pandas as pd
```

### Using pd.read_html() function to read HTML webpage

```python
df = pd.read_html(url, header = 0)
df
```

### How many tables are there in the webpage?

```python
len(df)
```

### Select the first table

```python
df[0]
```

```python
df2019 = df[0]
df2019
```

```python
df2019.drop('Genre', inplace = True, axis = 1)
df2019.drop('Running Time', inplace = True, axis = 1)
df2019.drop('Budget', inplace = True, axis = 1)
df2019
```

```python
df2019['Top 10 Gross'] = df2019['Top 10 Gross'].str.replace('\W', '')
df2019['Top 10 Gross'] = df2019['Top 10 Gross'].astype('int64')
df2019.info()
df2019
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df2019['Week'], df2019['Top 10 Gross'])
plt.title("2019 Top 10 Gross For Each Week")
plt.xlabel("Week")
plt.ylabel("Top 10 Gross (in $100 millions)")
plt.show()
# plotted a line graph with Week vs. Top 10 Gross amount
```

```python
# Save dataframe to csv

df2019.to_csv('data2019.csv')
```
