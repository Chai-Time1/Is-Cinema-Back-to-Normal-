```python
url2020 = "https://www.boxofficemojo.com/weekend/by-year/2020/"
url2020
```

```python
import pandas as pd
df = pd.read_html(url2020, header = 0)
df
```

```python
df2020 = df[0]
df[0]
```

```python
import seaborn as sns
```

```python
sns.lineplot(data = df2020, x = 'Week', y = 'Releases')
```

```python
sns.distplot(df2020.Releases, 
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2))
```

```python
sns.distplot(df2020.Releases,
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2),
             color='#00BFC4')
```
