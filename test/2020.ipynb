```python
url2020 = "https://www.boxofficemojo.com/weekend/by-year/2020/"
url2020
```

```python
import pandas as pd
df = pd.read_html(url2020, header = 0)
df
```

```python
df2020 = df[0]
df2020
```

```python
df2020.drop('Genre', inplace = True, axis = 1)
df2020.drop('Running Time', inplace = True, axis = 1)
df2020.drop('Budget', inplace = True, axis = 1)
df2020
```

```python
df2020['Top 10 Gross'] = df2020['Top 10 Gross'].str.replace('\W', '')
df2020['Top 10 Gross'] = df2020['Top 10 Gross'].astype('int64')
df2020.info()
df2020
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df2020['Week'], df2020['Top 10 Gross'])
plt.title("2020 Top 10 Gross For Each Week")
plt.xlabel("Week")
plt.ylabel("Top 10 Gross (in $100 millions)")
plt.show()
# plotted a line graph with Week vs. Top 10 Gross amount
```

```python
# Save dataframe to csv

df2020.to_csv('data2020.csv')
```
