```python
import pandas as pd
```

```python
df2015 = pd.read_csv('data2015.csv')




df2015 = df2015.drop(index=[5,6,13,19,35,43,52,56])

df2015 = df2015.reset_index()

df2015
```

```python
df2015_Top10 = df2015[['Week','Top 10 Gross']]
df2015_Top10.rename(columns = {'Week':'Week2015', 'Top 10 Gross':'Top102015'}, inplace = True)
df2015_Top10
```

```python
df2016 = pd.read_csv('data2016_real.csv')




df2016 = df2016.drop(index=[0,1,2,8,9,16,23,33,38,49,56,60])
df2016 = df2016.reset_index()

df2016.info()
```

```python
df2016_Top10 = df2016[['Week','Top 10 Gross']]
df2016_Top10.rename(columns = {'Week':'Week2016', 'Top 10 Gross':'Top102016'}, inplace = True)
```

```python
df2017 = pd.read_csv('data2017.csv')



df2017 = df2017.drop(index=[0,2,8,9,16,22,32,39,46,54,61])
df2017 = df2017.reset_index()
df2017.info()
```

```python
df2017_Top10 = df2017[['Week','Top 10 Gross']]
df2017_Top10.rename(columns = {'Week':'Week2017', 'Top 10 Gross':'Top102017'}, inplace = True)
```

```python
df2018 = pd.read_csv('data2018.csv')



df2018 = df2018.drop(index=[0,2,8,9,16,22,38,46,54,60])
df2018 = df2018.reset_index()

df2018.info()
```

```python
df2018_Top10 = df2018[['Week','Top 10 Gross']]
df2018_Top10.rename(columns = {'Week':'Week2018', 'Top 10 Gross':'Top102018'}, inplace = True)
```

```python
df2019 = pd.read_csv('data2019.csv')



df2019 = df2019.drop(index=[5,6,14,20,35,41,52,56])
df2019 = df2019.reset_index()
df2019.info()
```

```python
df2019_Top10 = df2019[['Week','Top 10 Gross']]
df2019_Top10.rename(columns = {'Week':'Week2019', 'Top 10 Gross':'Top102019'}, inplace = True)
```

```python
week_count = []
for i in range(52):
    week_count.append(52-i)
week_column = pd.DataFrame(week_count)
week_column.rename(columns = {0:'Week'}, inplace = True)
week_column
```

# Combine dfs together

```python
result = pd.concat([df2015_Top10['Top102015'], df2016_Top10['Top102016'], df2017_Top10['Top102017'], df2018_Top10['Top102018'], df2019_Top10['Top102019']], axis=1, join='inner')
# result.drop('Week2015', inplace = True,axis = 1)
# result.drop('Week2016', inplace = True,axis = 1)
# result.drop('Week2017', inplace = True,axis = 1)
# result.drop('Week2018', inplace = True,axis = 1)
# result.drop('Week2019', inplace = True,axis = 1)
result['mean'] = result.mean(axis=1)
result = pd.concat([result, week_column], axis = 1, join = 'inner')
result
```

```python
result.to_csv('Avgof2015to2019.csv')
```
